```python
'''
Figure 1: GMI Surface Maps
'''

import numpy as np
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter

mapfile = '/edata2/spencer/gpm_surf/GMI_surfmap_1712_V7.nc'

with xr.open_dataset(mapfile) as f:
    lat = f.lat.values
    lon = f.lon.values
    sfctype = f.SurfaceType[0,:,:].values
```

```python
from matplotlib.colors import ListedColormap, BoundaryNorm

fig = plt.figure(figsize=[10,8])
ax = fig.add_subplot(1,1,1, projection=ccrs.PlateCarree())


colors = {'1': 'midnightblue',
          '2': 'whitesmoke',
          '3': 'darkgreen',
          '4': 'forestgreen',
          '5': 'yellowgreen',
          '6': 'darkkhaki',
          '7': 'tan',
          '8': 'snow',
          '9': 'lightgrey',
          '10': 'darkgrey',
          '11': 'grey',
          '12': 'steelblue',
          '13': 'darkred',
          '14': 'firebrick',
          '15': 'gold',
          '16': 'aqua',
          '17': 'violet',
          '18': 'purple'}


values = np.arange(1, 19)
cmap = ListedColormap([colors[v] for v in colors.keys()])
norm = BoundaryNorm(np.arange(0.5, 18.5+1, 1), cmap.N)

mesh = ax.pcolormesh(lon,lat,sfctype, cmap=cmap, norm=norm)

# Add gridlines
gl = ax.gridlines(draw_labels=True, 
                  linewidth=0.8, color='gray', alpha=1., linestyle='--')
gl.top_labels = False
gl.right_labels = False
gl.xlabel_style = {'size': 11}
gl.ylabel_style = {'size': 11}

# Use proper lon/lat formatting
ax.xaxis.set_major_formatter(LongitudeFormatter(zero_direction_label=True))
ax.yaxis.set_major_formatter(LatitudeFormatter())

cb = plt.colorbar(mesh, ax=ax, ticks=values, orientation='vertical', shrink=0.5)
cb.set_label('Surface Type', fontsize=12)

ax.coastlines()
plt.show()
```

```python
'''
Figure: Errors
'''

import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import os
from contextlib import contextmanager
import xarray as xr
import importlib

@contextmanager
def cd(newdir):
    prevdir = os.getcwd()
    os.chdir(newdir)
    try:
        yield
    finally:
        os.chdir(prevdir)

gmi_path = '../GMI/GPM'
amsr2_path = '../AMSR2/GCOMW1'
ssmi_path = '../SSMI/F13'
ssmis_path = '../SSMIS/F17'
atms_path = '../ATMS/NOAA20'

with cd(gmi_path):
    import sensor_info
    importlib.reload(sensor_info)
    gmi_chans = sensor_info.channel_descriptions

with cd(amsr2_path):
    importlib.reload(sensor_info)
    amsr2_chans = sensor_info.channel_descriptions

with cd(ssmi_path):
    importlib.reload(sensor_info)
    ssmi_chans = sensor_info.channel_descriptions

with cd(ssmis_path):
    importlib.reload(sensor_info)
    ssmis_chans = sensor_info.channel_descriptions

with cd(atms_path):
    importlib.reload(sensor_info)
    atms_chans = sensor_info.channel_descriptions

def get_error_data(model_path, max_ocean = 5_000_000):

    with cd(model_path):
        import local_functions
        import model_operations
        import sensor_info
        importlib.reload(local_functions)
        importlib.reload(model_operations)
        importlib.reload(sensor_info)
        model_tree = model_operations.load_model_tree()
        #Ocean
        sfc = [1]
        with xr.open_dataset(f'training_data/{sensor_info.satellite}_training_data.nc') as f:
            sfctype = f.sfctype.values
            correct_sfc = sfctype == 1
            Tbs = f.Tbs.values[correct_sfc]
            sfctype = sfctype[correct_sfc]
        if Tbs.shape[0] > max_ocean:
            Tbs = Tbs[:max_ocean,:]
            sfctype = sfctype[:max_ocean]
        train_indcs, test_indcs, val_indcs = local_functions.split_data_indcs(Tbs)
        Tbs_test = Tbs[test_indcs]
        sfctype  = sfctype[test_indcs] 
        y_true = Tbs_test
        y_pred = model_operations.run_predictions(y_true, sfctype, model_tree, keep_dims=False)
        error_ocn = y_pred - y_true
        mean_error_ocn = np.mean(error_ocn, axis=0)
        std_error_ocn  = np.std(error_ocn, axis=0)
        
        #Nonocean
        with xr.open_dataset(f'training_data/{sensor_info.satellite}_training_data.nc') as f:
            sfctype = f.sfctype.values
            correct_sfc = sfctype > 1
            Tbs = f.Tbs.values[correct_sfc]
            sfctype = sfctype[correct_sfc]
        train_indcs, test_indcs, val_indcs = local_functions.split_data_indcs(Tbs)
        Tbs_test = Tbs[test_indcs]
        sfctype  = sfctype[test_indcs]
        y_true = Tbs_test
        y_pred = model_operations.run_predictions(y_true, sfctype, model_tree, keep_dims=False)
        error_non = y_pred - y_true
        mean_error_non = np.mean(error_non, axis=0)
        std_error_non  = np.std(error_non, axis=0)
    
    return error_ocn, error_non, mean_error_ocn, mean_error_non, std_error_ocn, std_error_non

(gmi_err_ocn, gmi_err_non, 
 gmi_mean_ocn, gmi_mean_non, 
 gmi_std_ocn, gmi_std_non) = get_error_data(gmi_path)
(amsr2_err_ocn, amsr2_err_non, 
 amsr2_mean_ocn, amsr2_mean_non, 
 amsr2_std_ocn, amsr2_std_non) = get_error_data(amsr2_path, max_ocean = 10_000_000)
(ssmi_err_ocn, ssmi_err_non, 
 ssmi_mean_ocn, ssmi_mean_non,
 ssmi_std_ocn, ssmi_std_non) = get_error_data(ssmi_path)
(ssmis_err_ocn, ssmis_err_non,
 ssmis_mean_ocn, ssmis_mean_non,
 ssmis_std_ocn, ssmis_std_non) = get_error_data(ssmis_path)
```

```python
def get_error_data_xt(model_path, scan_position, max_ocean=5_000_000):

    with cd(model_path):
        import local_functions
        import model_operations
        import sensor_info
        importlib.reload(local_functions)
        importlib.reload(model_operations)
        importlib.reload(sensor_info)
        model_tree = model_operations.load_model_tree()
        #Ocean
        sfc = [1]
        with xr.open_dataset(f'training_data/{sensor_info.satellite}_training_data.nc') as f:
            sfctype = f.sfctype.values
            correct_sfc = sfctype == 1
            Tbs = f.Tbs.values[correct_sfc]
            scanpos = f.scanpos.values[correct_sfc]
            sfctype = sfctype[correct_sfc]


        correct_scanpos = scanpos == scan_position
        Tbs = Tbs[correct_scanpos]
        sfctype = sfctype[correct_scanpos]
        scanpos = scanpos[correct_scanpos]

        train_indcs, test_indcs, val_indcs = local_functions.split_data_indcs(Tbs)
        Tbs_test = Tbs[test_indcs]
        sfctype  = sfctype[test_indcs]
        scanpos = scanpos[test_indcs]
        y_true = Tbs_test
        y_pred = model_operations.run_predictions(y_true, sfctype, scanpos, model_tree, keep_dims=False)
        error_ocn = y_pred - y_true
        mean_error_ocn = np.mean(error_ocn, axis=0)
        std_error_ocn  = np.std(error_ocn, axis=0)
        
        #Nonocean
        with xr.open_dataset(f'training_data/{sensor_info.satellite}_training_data.nc') as f:
            sfctype = f.sfctype.values
            correct_sfc = sfctype > 1
            Tbs = f.Tbs.values[correct_sfc]
            scanpos = f.scanpos.values[correct_sfc]
            sfctype = sfctype[correct_sfc]

        correct_scanpos = scanpos == scan_position
        Tbs = Tbs[correct_scanpos]
        sfctype = sfctype[correct_scanpos]
        scanpos = scanpos[correct_scanpos]
        
        train_indcs, test_indcs, val_indcs = local_functions.split_data_indcs(Tbs)
        Tbs_test = Tbs[test_indcs]
        sfctype  = sfctype[test_indcs]
        scanpos = scanpos[test_indcs]
        y_true = Tbs_test
        y_pred = model_operations.run_predictions(y_true, sfctype, scanpos, model_tree, keep_dims=False)
        error_non = y_pred - y_true
        mean_error_non = np.mean(error_non, axis=0)
        std_error_non  = np.std(error_non, axis=0)
    
    return error_ocn, error_non, mean_error_ocn, mean_error_non, std_error_ocn, std_error_non

(atms_err_ocn, atms_err_non,
 atms_mean_ocn, atms_mean_non,
 atms_std_ocn, atms_std_non) = get_error_data_xt(atms_path, scan_position=0, max_ocean=5_000_000)
```
