```python
'''
Test qc with no 85s
'''

import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

import local_functions
import model_operations
```

```python
test_file = '/edata1/archive/GPM/1C_F13_V7/0410/041004/1C.F13.SSMI.XCAL2018-V.20041004-S063548-E081737.049187.V07A.HDF5'

mlflag = model_operations.ml_quality_control(test_file, error_threshold=10)
print(np.sum(mlflag > 0, axis=0))
```

```python
'''
Plot the flag
'''

data = local_functions.read_ssmi_l1c(test_file)

fig = plt.figure(figsize=[12,10])
ax = fig.add_subplot(1,1,1, projection=ccrs.PlateCarree())
im = ax.scatter(data['lon'], data['lat'], c=mlflag[:,-1], transform=ccrs.PlateCarree(), s=0.3)
plt.colorbar(mappable=im, shrink=0.45)
plt.title('ML Quality Flag')
ax.coastlines()
plt.show()
```
