# Test run of GMI pretrained MLQC model

This notebook contains code to run the pretrained GMI MLQC model tree that is prepackaged with this release. This test version is set up to run on the test file found in the test data path `mlqc/data/l1c` by default. If the test data has not been downloaded, please download using the instructions in `mlqc/README.md`.

The first few cells below runs the premade MLQC function found in `mlqc/src/model_operations.py`. The next cell gives some example plots. The following cells manually walk through the steps of the MLQC process for instructional purposes, but the result is the same.

```python
'''
First run: test MLQC model operation
'''

import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import glob
import paths
from src import model_operations, sensor_info, surface
from src.utils import L1C
from matplotlib.colors import ListedColormap, BoundaryNorm
```

```python
#Get the file from the test data directory
l1cpath = paths.l1c_datapath
l1cfile = glob.glob(f'{l1cpath}/*.HDF5')[0]
mlqcflag = model_operations.ml_quality_control(l1cfile, error_threshold=4, keep_dims=True,
                                              ignore_l1cflag=True)
```

```python
#Read in lats, lons, and Tbs from Level 1C data for plotting
data = L1C.read_gmi_l1c(l1cfile)
```

```python
#Plot some Tbs and quality flags

cmap = ListedColormap(["gray", "red"])
norm = BoundaryNorm([0, 0.5, 1], cmap.N)

chan_desc = np.array(sensor_info.channel_descriptions)
chans2plot = ['10V', '89V', '166V']

nrows = len(chans2plot)

fig, axes = plt.subplots(nrows=nrows, ncols=2, 
                         figsize=[16,nrows*3.25], 
                         subplot_kw={'projection': ccrs.PlateCarree()})


axes = axes.flatten()

for i in range(nrows):
    
    chan = chans2plot[i]
    chan_indx = np.where(chan_desc == chan)[0][0]
    
    im1 = axes[i*2].scatter(data['lon'], 
                            data['lat'], 
                            c=data['Tbs'][:,:,chan_indx],
                            s=1.)
    
    im2 = axes[i*2+1].scatter(data['lon'],
                              data['lat'], 
                              c=mlqcflag[:,:,chan_indx],
                              s=1.,
                              cmap=cmap, norm=norm)
    
    plt.colorbar(mappable=im1, ax=axes[i*2], shrink=0.7,pad=0.05)
    cbar = plt.colorbar(mappable=im2, ax=axes[i*2+1], shrink=0.7, pad=0.05,
                 ticks=[0, 1])
    cbar.ax.set_yticklabels(["0", "1"])
    
    gl = axes[i*2].gridlines(draw_labels=False)
    gl = axes[i*2+1].gridlines(draw_labels=False)
    
    axes[i*2].coastlines()
    axes[i*2+1].coastlines()
    axes[i*2].set_title(f'{chan} Tb [K]')
    axes[i*2+1].set_title(f'{chan} MLQC Flag Value')


plt.show()

```

```python
'''
Now, walk through the process step-by-step
'''

error_threshold = 4

#Read the Level 1C File
data = L1C.read_gmi_l1c(l1cfile)

#Get dims before reshaping
nscans, npixs, nchans = data['Tbs'].shape 

#Reshape data to be fed into model
Tbs      = data['Tbs'].reshape(-1, nchans)   
lat      = data['lat'].reshape(-1)   
lon      = data['lon'].reshape(-1)  
scantime = np.repeat(data['scantime'], npixs)  
qual     = data['qual'].reshape(-1,sensor_info.nscangroups)   

#Now get length of sampling dimension
nsamples = Tbs.shape[0]  

#Create initial qc flag    
mlqcflag = np.full(Tbs.shape, dtype=np.int32, fill_value=-99)  

#Create internal quality flag (helps keep shapes the same)
internal_qual = np.zeros(nsamples, dtype=np.byte) 

#Check for any remaining NaNs in L1C 
#(have to get rid of them before trying to attach surface types)
nans = np.logical_or(np.isnan(lat), np.isnan(lon)) 
internal_qual[nans] = 1   
nans = np.any(np.isnan(Tbs), axis=1) 
internal_qual[nans] = 1   

good = internal_qual == 0  

#Attach surface type to good pixels   
sfctype = np.full([nsamples], dtype=np.int32, fill_value=-99)  
sfctype[good] = surface.attach_gpm_sfctype(lat[good],
                                           lon[good],
                                           scantime[good],
                                           sensor=sensor_info.sensor)



#Load in model tree
model_tree = model_operations.load_model_tree() 

#Predict Tbs      
Tbs_pred = np.full(Tbs.shape, dtype=Tbs.dtype, fill_value=np.nan)
Tbs_pred[good] = model_operations.run_predictions(Tbs[good],
                                                  sfctype[good], 
                                                  model_tree, 
                                                  keep_dims=False)  

#Calculate standardized prediction errors 
epsilon_hat = np.full(Tbs.shape, dtype=Tbs.dtype, fill_value=np.nan)
epsilon_hat[good] = model_operations.get_error(Tbs[good], 
                                               Tbs_pred[good], 
                                               sfctype[good])  

#Set flag
mlqcflag[epsilon_hat > error_threshold] = 1 
mlqcflag[epsilon_hat <= error_threshold] = 0   

mlqcflag = mlqcflag.reshape(nscans,npixs,nchans)
```

```python
#Plot some Tbs and quality flags

cmap = ListedColormap(["gray", "red"])
norm = BoundaryNorm([0, 0.5, 1], cmap.N)

chan_desc = np.array(sensor_info.channel_descriptions)
chans2plot = ['10V', '89V', '166V']

nrows = len(chans2plot)

fig, axes = plt.subplots(nrows=nrows, ncols=2, 
                         figsize=[16,nrows*3.25], 
                         subplot_kw={'projection': ccrs.PlateCarree()})


axes = axes.flatten()

for i in range(nrows):
    
    chan = chans2plot[i]
    chan_indx = np.where(chan_desc == chan)[0][0]
    
    im1 = axes[i*2].scatter(data['lon'], 
                            data['lat'], 
                            c=data['Tbs'][:,:,chan_indx],
                            s=1.)
    
    im2 = axes[i*2+1].scatter(data['lon'],
                              data['lat'], 
                              c=epsilon_hat[:,chan_indx],
                              s=1.,
                              cmap='plasma')
    
    plt.colorbar(mappable=im1, ax=axes[i*2], shrink=0.7,pad=0.05)
    cbar = plt.colorbar(mappable=im2, ax=axes[i*2+1], shrink=0.7, pad=0.05)
    
    gl = axes[i*2].gridlines(draw_labels=False)
    gl = axes[i*2+1].gridlines(draw_labels=False)
    
    axes[i*2].coastlines()
    axes[i*2+1].coastlines()
    axes[i*2].set_title(f'{chan} Tb [K]')
    axes[i*2+1].set_title(f'{chan} Absolute Prediction Error $(\sigma)$')


plt.show()

```
